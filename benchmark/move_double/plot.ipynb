### Difference between `move_double` on and off

We use black line to denote the reference result which `double_insert` and `double_remove` are disabled. 

```python
from triqs.gf import *
from triqs.plot.mpl_interface import oplot

from h5 import *

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
cmap = plt.colormaps["hsv"]
%config InlineBackend.figure_format = 'retina'
mpl.rcParams['figure.dpi'] = 150
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['mathtext.rm'] = 'serif'
plt.rc('font', size=12, family='Times New Roman')

def normalize_color(color):
    brightness = color[0]**2 + color[1]**2 + color[2]**2
    return (color[0]/brightness, color[1]/brightness, color[2]/brightness, color[3])
```

- Multiorbital

```python
with HDFArchive("multiorb.ref.h5", "r") as A:
    G_tau_ref = A["G_tau"]

with HDFArchive("multiorb.out.h5", "r") as A:
    G_tau_out = A["G_tau"]

oplot(G_tau_ref["up1"].real, 'k', linewidth = 2, label="Ref")
oplot(G_tau_out["up1"].real, color="aqua", linewidth = 0.6, label="Move double")
plt.title("Multiorbital (3)")
plt.legend()
plt.show()
```

- Spin-Spin

```python
with HDFArchive("spin_spin.ref.h5", "r") as A:
    G_tau_ref = A["G_tau"]

with HDFArchive("spin_spin.out.h5", "r") as A:
    G_tau_out = A["G_tau"]

oplot(G_tau_ref["up"].real, 'k', linewidth = 2, label="Ref")
oplot(G_tau_out["up"].real, color="aqua", linewidth = 0.6, label="Move double")
plt.title("Spin-Spin")
plt.legend()
plt.show()
```
