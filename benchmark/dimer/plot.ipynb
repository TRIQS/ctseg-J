### Compare `dimer.py` result with exact diagonolization

We use black line to denote the ED result. 

```python
from triqs.gf import *
from triqs.plot.mpl_interface import oplot

from h5 import *

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
cmap = plt.colormaps["hsv"]
%config InlineBackend.figure_format = 'retina'
mpl.rcParams['figure.dpi'] = 150
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['mathtext.rm'] = 'serif'
plt.rc('font', size=12, family='Times New Roman')

def normalize_color(color):
    brightness = color[0]**2 + color[1]**2 + color[2]**2
    return (color[0]/brightness, color[1]/brightness, color[2]/brightness, color[3])
```

```python
with HDFArchive("dimer_pyed.ref.h5", "r") as A:
    G_tau_ed = A["G_tau"]

with HDFArchive("dimer.out.h5", "r") as A:
    G_tau_seg = A["G_tau"]

oplot(G_tau_ed["up"][0, 0].real, 'k', linewidth = 2, label="ED")
oplot(G_tau_ed["up"][0, 1].real, 'k', linewidth = 2, label="ED")
oplot(G_tau_seg["up"][0, 0].real, color="aqua", linewidth = 0.6, label="ctseg")
oplot(G_tau_seg["up"][0, 1].real, color="aqua", linewidth = 0.6, label="ctseg")
plt.title("Dimer")
plt.ylabel(r"$G(\tau)$")
plt.legend()
plt.show()
```
