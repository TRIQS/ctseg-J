### Compare `spin_spin.py` result with `ctint`

We use black line to denote the ctint result. 

```python
from triqs.gf import *
from triqs.plot.mpl_interface import oplot

from h5 import *

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
cmap = plt.colormaps["hsv"]
%config InlineBackend.figure_format = 'retina'
mpl.rcParams['figure.dpi'] = 150
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['mathtext.rm'] = 'serif'
plt.rc('font', size=12, family='Times New Roman')

def normalize_color(color):
    brightness = color[0]**2 + color[1]**2 + color[2]**2
    return (color[0]/brightness, color[1]/brightness, color[2]/brightness, color[3])
```

```python
with HDFArchive("ctint.ref.h5", "r") as A:
    G_tau_int = A["dmft_loop/i_001/S/G_tau"]

with HDFArchive("spin_spin.ref.h5", "r") as A:
    G_tau_seg = A["G_tau"]

oplot(G_tau_int["up"].real, 'k', linewidth = 2, label="ctint")
oplot(G_tau_seg["up"].real, color="aqua", linewidth = 0.6, label="ctseg")
plt.title("Spin Spin")
plt.ylabel(r"$G(\tau)$")
plt.legend()
plt.show()
```
